```python

# Configure
c = twint.Config()
c.Search = "bayonetta AND smash"
# c.Search = "bayonetta AND broken OR op OR overpowered OR busted OR easy OR lame OR lamest OR dumb OR stupid OR boring OR ban OR banned OR nerf" 
# This query looks a little intense ^^^ The ORs will get a lot of noise so you might consider breaking these down into separate queries
c.Pandas = True
c.Lang= "en"
c.Since = "2016-02-03 00:00:00"
# c.Since = "2016-2-3"
# c.Until = "2018-12-8"
c.Hide_output = True

# Run
twint.run.Search(c)
bayo_df = twint.storage.panda.Tweets_df
```

```python
bayo_df.sort_values(by='date')
```

```python
#Filter to only English tweets
bayo_df = bayo_df[bayo_df['language'] == 'en']
```

```python
bayo_df
```

```python
#Filter by date to before Smash Ultimate released to only get tweets about Smash 4 Bayonetta
bayo_df_smash4 = bayo_df[bayo_df['date'] <= '2018-12-07 00:00:00']
```

```python
bayo_df_smash4
```

```python
#Save to CSV
bayo_df_smash4.to_csv('bayo.csv')
```
